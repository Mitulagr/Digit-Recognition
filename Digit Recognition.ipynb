```python
#=================INPUT=================

ln = 5000 
# Length of the Training Set (Elements with index 0 to ln-1 will be taken)

div = 50 
# Number of divisions of the Training Set
# Note : Keep (ln/div) more than 100 otherwise there might be error due to non-occurance of a digit in a group

start = 10000 
# Starting index of the Test Set

end = 10100
# Ending index of the Test Set = end-1
# Note : There are a total 42000 elements available in the set

p = 2
# Power of the difference between pixel value of input and average of each group for each digit

knn = 3
# Number of lowest deviations to consider for final output

#=======================================
```

```python
import numpy as np
import pandas as pd
import math

x = pd.read_csv('train.csv')
y = x["label"]
```

```python
a0l=[0]*div
a1l=[0]*div
a2l=[0]*div
a3l=[0]*div
a4l=[0]*div
a5l=[0]*div
a6l=[0]*div
a7l=[0]*div
a8l=[0]*div
a9l=[0]*div

a0 = [0]*784
a1 = [0]*784
a2 = [0]*784
a3 = [0]*784
a4 = [0]*784
a5 = [0]*784
a6 = [0]*784
a7 = [0]*784
a8 = [0]*784
a9 = [0]*784

for i in range(784):
    a0[i] = [0]*div
    a1[i] = [0]*div
    a2[i] = [0]*div
    a3[i] = [0]*div
    a4[i] = [0]*div
    a5[i] = [0]*div
    a6[i] = [0]*div
    a7[i] = [0]*div
    a8[i] = [0]*div
    a9[i] = [0]*div


for i in range (ln) :
    k = math.floor((div*i)/ln)
    if(y[i]==0) : 
        a0l[k] = a0l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a0[j][k] = a0[j][k] + int(x[u][i])  
    if(y[i]==1) : 
        a1l[k] = a1l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a1[j][k] = a1[j][k] + int(x[u][i])  
    if(y[i]==2) : 
        a2l[k] = a2l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a2[j][k] = a2[j][k] + int(x[u][i])  
    if(y[i]==3) : 
        a3l[k] = a3l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a3[j][k] = a3[j][k] + int(x[u][i])   
    if(y[i]==4) : 
        a4l[k] = a4l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a4[j][k] = a4[j][k] + int(x[u][i])  
    if(y[i]==5) : 
        a5l[k] = a5l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a5[j][k] = a5[j][k] + int(x[u][i])  
    if(y[i]==6) : 
        a6l[k] = a6l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a6[j][k] = a6[j][k] + int(x[u][i])  
    if(y[i]==7) : 
        a7l[k] = a7l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a7[j][k] = a7[j][k] + int(x[u][i])  
    if(y[i]==8) : 
        a8l[k] = a8l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a8[j][k] = a8[j][k] + int(x[u][i])  
    if(y[i]==9) : 
        a9l[k] = a9l[k] + 1
        for j in range (784) : 
            u = "pixel"+str(j)
            a9[j][k] = a9[j][k] + int(x[u][i])              

for i in range (784) : 
    for k in range (div) :  
        a0[i][k] = a0[i][k]/a0l[k]
        a1[i][k] = a1[i][k]/a1l[k]
        a2[i][k] = a2[i][k]/a2l[k]
        a3[i][k] = a3[i][k]/a3l[k]
        a4[i][k] = a4[i][k]/a4l[k]
        a5[i][k] = a5[i][k]/a5l[k]
        a6[i][k] = a6[i][k]/a6l[k]
        a7[i][k] = a7[i][k]/a7l[k]
        a8[i][k] = a8[i][k]/a8l[k]
        a9[i][k] = a9[i][k]/a9l[k]       
```

```python
chc = [0]*10
ans = 0
anv = 0
acc = 0 

dev = [0]*10
for i in range(10):
    dev[i] = [0]*div

dig = [0]*(div*10)
for i in range (10) : 
    for v in range (div) : 
        dig[v+(i*div)] = i

for i in range (start,end) :  
    ans = 0
    anv = 0
    chc = [0]*10
    dev = [0]*10
    for v in range(10):
        dev[v] = [0]*div
    dig = [0]*(div*10)
    for m in range (10) : 
        for v in range (div) : 
            dig[v+(m*div)] = m
            
    for j in range (784) : 
        u = "pixel"+str(j)
        for k in range (div) : 
            dev[0][k] = dev[0][k] + abs((int(x[u][i]))-a0[j][k])**p
            dev[1][k] = dev[1][k] + abs((int(x[u][i]))-a1[j][k])**p
            dev[2][k] = dev[2][k] + abs((int(x[u][i]))-a2[j][k])**p
            dev[3][k] = dev[3][k] + abs((int(x[u][i]))-a3[j][k])**p
            dev[4][k] = dev[4][k] + abs((int(x[u][i]))-a4[j][k])**p
            dev[5][k] = dev[5][k] + abs((int(x[u][i]))-a5[j][k])**p
            dev[6][k] = dev[6][k] + abs((int(x[u][i]))-a6[j][k])**p
            dev[7][k] = dev[7][k] + abs((int(x[u][i]))-a7[j][k])**p
            dev[8][k] = dev[8][k] + abs((int(x[u][i]))-a8[j][k])**p
            dev[9][k] = dev[9][k] + abs((int(x[u][i]))-a9[j][k])**p
    
    dev = np.array(dev).reshape(10*div)
    zp = zip(dev, dig)
    z = [dig for _, dig in sorted(zp)]
    
    for k in range (knn) :
        chc[int(z[k])] = chc[int(z[k])] + 1
        
    ans = z[0]
    anv = chc[ans]
    for k in range (10) : 
        if(chc[k]>anv) :
            anv = chc[k]
            ans = k    
            
    if(int(ans) == int(y[i])) : 
        acc = acc + 1             

acc = acc/(end - start)
print(acc)
```
